```python
import cv2 
import matplotlib.pyplot as plt
import glob
import os
```

```python
path_images = "./carvan_dataset/train_images/"
path_masks = "./carvan_dataset/train_masks/"

img_paths = glob.glob(os.path.join(path_images, "*.*"))
img_paths = sorted(img_paths)
img_paths
```

```python
mask_paths = glob.glob(os.path.join(path_masks, "*.*"))
mask_paths = sorted(mask_paths)
mask_paths
```

```python
def new_path(p):
    fname = ""
    for c in p[::-1]:
        if c == '/' :
            break
        else:
            fname += c

    fname = fname[::-1]

    return "./carvan_dataset/train_mask2/" + fname 

print(new_path(img_paths[0]))
```

```python
tot = len(mask_paths)
done = 0
for p in mask_paths:
    img = cv2.imread(p)
    img[img>0] = 255 
    cv2.imwrite(new_path(p), img)
    done += 1
    if done % 10 == 0 : 
        print("Done ", done, " out of ", tot, "images.")
```
